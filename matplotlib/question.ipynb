```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Define the range of x values
x = np.linspace(-10, 10, 400)

# Linear Function: y = mx + c
m, c = 2, 3
y_linear = m * x + c

# Quadratic Function: y = ax^2 + bx + c
a, b, c = 1, -2, 3
y_quadratic = a * x**2 + b * x + c

# Sine and Cosine Functions
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot Linear Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_linear, label='y = 2x + 3', color='r')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
# plt.show()

# Plot Quadratic Function
plt.figure(figsize=(6, 4))
plt.plot(x, y_quadratic, label='y = x² - 2x + 3', color='g')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
# plt.show()

# Plot Sine & Cosine Functions
plt.figure(figsize=(6, 4))
plt.plot(x, y_sin, label='sin(x)', color='b')
plt.plot(x, y_cos, label='cos(x)', color='m')
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
