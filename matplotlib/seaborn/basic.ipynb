```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
#relational plot: scatterplotting(using axis level function)
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue= 'sex', size = 'size', style  = 'time')
```

```python
#figure level function:
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue= 'sex', size = 'size', style  = 'time', kind = 'scatter')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue= 'sex', size = 'size', style  = 'time', kind = 'line')
```

```python
import plotly.express as px
gap = px.data.gapminder()
```

```python
gap.head()
```

```python
temp_df = gap[gap['country']=='India']
```

```python
temp_df
```

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

```python
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', kind = 'line')
```

```python
temp_data = gap[gap['country'].isin(['India','Pakistan','China', 'Brazil','Germany'])]
```

```python
temp_data
```

```python
sns.relplot(data = temp_data, x = 'year', y = 'lifeExp', kind = 'line', hue = 'country')
sns.relplot(data = temp_data, x = 'year', y = 'gdpPercap', kind = 'line', hue = 'country')
```

```python
sns.relplot(data = temp_data, x = 'year', y = 'lifeExp', kind = 'line',hue = 'country', style= 'continent',size =  'continent')
```

```python
#facet plot: figure level using relplot()
# do not work with sdcatterplot or lineplot
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip',col = 'sex', row  = 'day')
```

```python
sns.relplot(data = gap, x = 'lifeExp', y = 'gdpPercap',col = 'year', col_wrap=3)
```
