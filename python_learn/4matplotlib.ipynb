```python
data = {
    "Carname": ["BMW", "Volvo", "VW", "VW", "Ford", "VW", "Tesla", "BMW", "Volvo", "Ford", "Toyota", "VW", "Toyota"],
    "Color": ["red", "black", "gray", "white", "white", "white", "red", "black", "gray", "white", "gray", "white", "blue"],
    "Age": [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6],
    "Speed": [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86],
    "AutoPass": ["Y", "Y", "N", "Y", "Y", "Y", "Y", "Y", "N", "N", "N", "N", "Y"]
}
```

```python
df = pd.DataFrame(data)

# Save to CSV for further processing
df.to_csv("car_dataset.csv", index = False)      #to_csv(csv_file_name)

# Display the dataset
print(df)
```

```python
'''
plt.plot(Carname, Speed, label= 'Age', color = "b", marker="o", linestyle = '_')
plt.xlable ("Carname")
plt.ylable ("Speed")
plt.title("Brand Speed")
plt.legend()
plt.grid(True)
plt.show()'''
import matplotlib.pyplot as plt
import pandas as pd

# Plot
plt.plot(df["Carname"], df["Speed"], label='Speed', color="y", marker="o", linestyle="--")

plt.xlabel("Car Name")  
plt.ylabel("Speed")  

plt.title("Brand Speed")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# 📊 **1. Bar Plot: Car Name vs. Speed**
plt.figure(figsize=(8, 4))
plt.bar(df["Carname"], df["Speed"], color="skyblue", edgecolor="black")
plt.xlabel("Car Name")
plt.ylabel("Speed")
plt.title("Bar Plot: Car Name vs. Speed")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# 📈 **2. Scatter Plot: Age vs. Speed**
plt.figure(figsize=(6, 4))
plt.scatter(df["Age"], df["Speed"], color="red", alpha=0.6)
plt.xlabel("Age of Car")
plt.ylabel("Speed")
plt.title("Scatter Plot: Age vs. Speed")
plt.grid(True)
plt.show()
```

```python
# 🥧 **3. Pie Chart: Distribution of Car Colors**
color_counts = df["Color"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(color_counts, labels=color_counts.index, autopct="%1.1f%%", colors=["red", "black", "gray", "white", "blue"], startangle=90)
plt.title("Pie Chart: Car Color Distribution")
plt.show()
```

```python
# 📊 **4. Histogram: Speed Distribution**
plt.figure(figsize=(6, 4))
plt.hist(df["Speed"], bins=5, color="green", edgecolor="black", alpha=0.7)
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.title("Histogram: Speed Distribution")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python

# 📦 **5. Box Plot: Speed Variation by Car Name**
plt.figure(figsize=(8, 5))
df.boxplot(column="Speed", by="Carname", grid=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.xlabel("Car Name")
plt.ylabel("Speed")
plt.title("Box Plot: Speed Variation by Car Name")
plt.suptitle("")  # Remove default title
plt.xticks(rotation=45)
plt.show()
```
