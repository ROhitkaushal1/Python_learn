```python
'''Shiwam Kumar'''
'''Assignment_1 (Milestone_Project_1): Income-tax-Calculator'''
'''
The program initializes an empty library (a dictionary with a list of books
Various functions handle different tasks:
add_book(): Adds a book with title, author, and availability.
remove_book(): Removes a book by title.
borrow_book(): Marks a book as borrowed (if available).
return_book(): Marks a book as available again.
display_books(): Shows all books and their statuses.

The main() function provides a menu-driven interface, allowing users to interact with the system.'''
import json
import os

class Library:
    def __init__(self,filename="library_data.json"):
        self.filename = filename
        self.books = self.load_books()

    def load_books(self):
        """Load books from the JSON file (if exists)."""
        if os.path.exists(self.filename):
            with open(self.filename, "r") as file:
                return json.load(file)  # Load existing book data
        return {}  # Return empty dictionary if file doesn't exist

    def save_books(self):
        """Save the current book list to the JSON file."""
        with open(self.filename, "w") as file:
            json.dump(self.books, file, indent=4)  # Save books with formatting

    def add_book(self, title, author):
        if title in self.books:
            print(f" '{title}' is already in the library!")
        else:
            self.books[title] = {"author": author, "available": True}
            print(f" Book '{title}' by {author} added to the library.")

    def remove_book(self, title):
        if title in self.books:
            del self.books[title]
            print(f" Book '{title}' removed from the library.")
        else:
            print(f" Book '{title}' not found in the library.")

    def borrow_book(self, title):
        if title in self.books:
            if self.books[title]["available"]:
                self.books[title]["available"] = False
                print(f" You have borrowed '{title}'. Enjoy reading!")
            else:
                print(f" Sorry, '{title}' is already borrowed.")
        else:
            print(f" Book '{title}' not found in the library.")

    def return_book(self, title):
        if title in self.books:
            if not self.books[title]["available"]:
                self.books[title]["available"] = True
                print(f" Thank you for returning '{title}'!")
            else:
                print(f" '{title}' was not borrowed.")
        else:
            print(f" Book '{title}' not found in the library.")

    def display_books(self):
        if not self.books:
            print("No books available in the library.")
        else:
            print("\n Available Books in the Library:")
            for title, info in self.books.items():
                status = " Available" if info["available"] else " Borrowed"
                print(f" '{title}' by {info['author']} - {status}")

def main():
    library = Library()

    while True:
        print("\nüìå Library Management System")
        print("1Ô∏è‚É£ Add Book")
        print("2Ô∏è‚É£ Remove Book")
        print("3Ô∏è‚É£ Borrow Book")
        print("4Ô∏è‚É£ Return Book")
        print("5Ô∏è‚É£ Display Books")
        print("6Ô∏è‚É£ Exit")
        
        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            title = input("Enter book title: ").strip()
            author = input("Enter author name: ").strip()
            library.add_book(title, author)
        elif choice == "2":
            title = input("Enter book title to remove: ").strip()
            library.remove_book(title)
        elif choice == "3":
            title = input("Enter book title to borrow: ").strip()
            library.borrow_book(title)
        elif choice == "4":
            title = input("Enter book title to return: ").strip()
            library.return_book(title)
        elif choice == "5":
            library.display_books()
        elif choice == "6":
            print("Exiting Library System. Have a great day!")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()


        
    
```
